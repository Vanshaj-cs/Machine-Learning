```python
import numpy as np
```

```python
class SVM_classifier():
  def __init__ (self, learning_rate, no_of_iterations, lambda_parameter):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations
    self.lambda_parameter = lambda_parameter

  def fit (self, X, Y):

    self.m, self.n = X.shape

    self.w = np.zeros(self.n)
    self.b = 0

    self.X = X;
    self.Y = Y

    for i in range (self.no_of_iterations):
      self.update_weights()

  def update_weights(self):
    y_label = np.where(self.Y <= 0, -1, 1)

    for index, x_i in enumerate(self.X):
      condition = y_label[index]*( np.dot (x_i, self.w) - self.b) >= 1

      if(condition == True):
        dw = 2 * self.lambda_parameter * self.w
        db = 0
      else:
        dw = (2 * self.lambda_parameter * self.w) - np.dot(x_i, y_label[index])
        db = y_label[index]

      self.w = self.w - self.learning_rate * dw
      self.b = self.b - self.learning_rate * db

  def predict(self, X):
    output = np.dot(X, self.w) - self.b
    predicted_labels = np.sign(output)

    y_hat = np.where(predicted_labels <= -1, 0, 1)

    return y_hat
```

Implementation

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
diabetes = pd.read_csv('diabetes.csv')
```

```python
features = diabetes.drop(columns = 'Outcome', axis = 1)
target = diabetes['Outcome']
```

Standardization

```python
scaler = StandardScaler()
scaler.fit(features)
standardized_data = scaler.transform(features)
```

Spliting

```python
X_train, X_test, Y_train, Y_test = train_test_split(standardized_data, target, test_size = 0.2, random_state = 2)
```

Training model

```python
classifier = SVM_classifier(learning_rate = 0.001, no_of_iterations = 1000, lambda_parameter = 0.01)
```

```python
classifier.fit(X_train, Y_train)
```

Model Evalutation

```python
training_data_pred = classifier.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, training_data_pred)
print(training_data_accuracy)
```

```python
test_data_accuracy = classifier.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, test_data_accuracy)
print(test_data_accuracy)
```
